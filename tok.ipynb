# Building a Tokenizer 

Using a WordPiece approach, building three tokenizers from reasonably small corpora.

```python
#import files from Huggining Face
from tokenizers import models, normalizers, pre_tokenizers, trainers, Tokenizer
from pathlib import Path
```

### Normalization Function

An initial clean-up and transformation stage for getting data ready to be tokenized.
<li>Remove the unnecessary white space in the data.</li>
<li>Convert the text to lower case in the data.</li>

```python
def normalization(text): 
    #remove unnecessary white space
    text = text.strip()

    #convert text to lower case
    text = text.lower()
    
    return text
```

### Pre-tokenization Function

Splits initial text into smaller units based on spaces.

```python
def pretokenization(tokenizer):
    #create the object for pre-tokenization operation
    tokenizer.pre_tokenizer = pre_tokenizers.BertPreTokenizer()
    
    return tokenizer
```

### Training Tokenizer Function

Trains a tokenizer using the training data set.

```python
def training(tokenizer, pre_tokens):
    #create the object for training operation
    trainer = trainers.WordPieceTrainer(vocab_size = 25000, special_tokens = ["[UNK]"])

    #use the trainer operation on the pre-tokenized data
    tokenizer.train_from_iterator(pre_tokens, trainer = trainer)
    
    return tokenizer
```

### Creating the Tokenizer

Creates a tokenizer object to be modified. Also modifies the tokenizer to normalize data such that unicode is uniform an all accents are stripped.

```python
def create_tokenizer():
    #create tokenizer object for WordPiece method
    tokenizer = Tokenizer(models.WordPiece(unk_token = "[UNK]"))
    tokenizer.normalizer = normalizers.Sequence([normalizers.NFD(), normalizers.StripAccents()]
)
    return tokenizer
```

## English Tokenizer

### Load Data Set

For the code to work, the data set needs to be in a folder labeled "English" in the location of this file.

```python
#initialize folder location
folder_path = Path("English")

#initialize container for text data
text = []

#iterate through file location where dataset is
for file_path in folder_path.iterdir():
    #check if location is a file
    if file_path.is_file():

        #open file and extract text from file
        with open(file_path, 'r', encoding = 'utf-8') as file:
            content = file.read()
            
            #normalize the text in the file
            text.append(content)
```

### Implement English Tokenizer

```python
#create tokenizer object
tokenizer = create_tokenizer()

#iterate through sections of text
for i in range(len(text)):    
    #normalize section of text
    text[i] = normalization(text[i])

#pretokenize data
tokenizer = pretokenization(tokenizer)

#train data
tokenizer = training(tokenizer, text)
```

### Encode Sample

```python
#initialize sample text
sample = """In recent years, fashion has been strongly influenced by nostalgia, with styles from the nineteen-nineties and early two-thousands
making a notable return. Items such as wide-leg jeans, thick-soled sneakers, butterfly hair clips, and midriff-baring tops have become popular
once again, creating a sense that the fashion of those decades has re-emerged. Celebrities like Dua Lipa and Bella Hadid have contributed to the
revival by wearing ensembles reminiscent of those seen in films like Clueless or inspired by Bratz dolls. Many well-known brands are also 
embracing this trend by reintroducing previous collections or partnering with prominent figures from past decades. This demonstrates how, in
popular culture, past trends often find new life in the present."""

#use trained tokenizer to tokenize the paragraph
encoding = tokenizer.encode(sample)
print(encoding.tokens)
```
